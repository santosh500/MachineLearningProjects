```python
import numpy as np
import pandas as pd
import os as os
import cv2
import shutil
from PIL import Image, ImageDraw
```

```python
#2 THE MAIN PART (Resize the images and migrate them to actual folders)
for files in os.listdir('./train'):
    str1 = str('./train') + files
    print(str1)
    for files2 in os.listdir('./train/' + files):
        print(files2)
        img = cv2.imread('./train/' + files + '/' + files2, cv2.IMREAD_UNCHANGED)
        print('Original Dimensions : ',img.shape)
        width = 128
        height = 128
        #height = img.shape[0] # keep original height
        dim = (width, height)
        # resize image
        resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        print('Original Dimensions : ',resized.shape)
        o = files2.split('_')
        cv2.imwrite('./trainMain/'+o[0] + '/' +files2,resized)
```
